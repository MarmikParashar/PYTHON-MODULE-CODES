```python
# merge the image horizentally
#to perform the merge operation, we need to make sure both images should be the same size.

from PIL import Image

# Load the images
image1 = Image.open("C:\\Users\\marmi\\OneDrive\\Pictures\\Saved Pictures\\wp 1.jpg")
image1.show()
```

```python
image2 = Image.open("C:\\Users\\marmi\\OneDrive\\Pictures\\Saved Pictures\\wp 2.jpg")
image2.show()
```

```python
# Get the dimensions of the images
width1, height1 = image1.size
width1, height1
```

```python
width2, height2 = image2.size
width2, height2
```

```python
# Create a new image with the combined width and the height of the tallest image
new_width = width1 + width2
new_height = max(height1, height2)
new_image = Image.new("RGB", (new_width, new_height))
```

```python

# Paste the two images onto the new image
new_image.paste(image1, (0, 0))
new_image.paste(image2, (width1, 0))
new_image.show()
# Save the new image
#new_image.save("C:\\Users\\marmi\\OneDrive\\Pictures\\Saved Pictures\\newImage.jpg")
```

```python
# merge vertically

from PIL import Image

# Load the images
image1 = Image.open("C:\\Users\\marmi\\OneDrive\\Pictures\\Saved Pictures\\wp 1.jpg")
image2 = Image.open("C:\\Users\\marmi\\OneDrive\\Pictures\\Saved Pictures\\wp 2.jpg")

# Get the dimensions of the images
width1, height1 = image1.size
print("Dimensions of Image 1 :",width1, height1)
width2, height2 = image2.size
print("Dimensions of Image 2 :",width2, height2)
```

```python
# Create a new image with the combined width and the height of the tallest image
new_width = max(width1, width2)
print("New Width :" ,new_width)
new_height = height1 + height2
print("New Height :",new_height)
new_image = Image.new("RGB", (new_width, new_height))
print("New Image :",new_image)
```

```python
# Paste the two images onto the new image
new_image.paste(image1, (0, 0)) 
new_image.paste(image2, (0, height1))
new_image.show()

# Save the new image
#new_image.save("E:/data/img54.jpg","JPEG",quality=57)
#new_image.show()
```

```python
# Enhance Brightness of image
from PIL import Image
from PIL import ImageEnhance 
image = Image.open("C:\\Users\\marmi\\OneDrive\\Pictures\\Saved Pictures\\wp 1.jpg")
image.show() 
```

```python
curnt_bri = ImageEnhance.Brightness(image)
curnt_bri
```

```python
new_bri = 1.5

# Brightness enhanced by  1.5 
af_brt = curnt_bri.enhance(new_bri) 
  
# shows updated image in image viewer 
af_brt.show()  
```

```python
#Enhance Contrast of image

from PIL import Image 
from PIL import ImageEnhance 
  
# Opens the image file 
image = Image.open("C:\\Users\\marmi\\OneDrive\\Pictures\\Saved Pictures\\wp 1.jpg")
image.show()
```
